```python
from PIL import Image
import glob
import os

sourcePath = './datasets/customizedImg/sourceImg/png/'
targetPath = './datasets/customizedImg/sourceImg/'

img_lst = glob.glob(sourcePath + '*.png')

for f in img_lst:
    
    img = Image.open(f)
    background = Image.new("RGB", img.size, (255,255,255))
    background.paste(img, (0,0), img)
    background.save(f[:-3]+'jpg', quality=95)
```
